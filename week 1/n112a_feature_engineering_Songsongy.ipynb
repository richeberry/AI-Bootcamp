```python
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g_0.csv'
headers = ['ë¶„ê¸°', 'ë§¤ì¶œì•¡', 'ì˜ì—…ì´ìµ', 'ì˜ì—…ì´ìµ(ë°œí‘œê¸°ì¤€)', 'ì„¸ì „ê³„ì†ì‚¬ì—…ì´ìµ',
           'ë‹¹ê¸°ìˆœì´ìµ', 'ë‹¹ê¸°ìˆœì´ìµ(ì§€ë°°)', 'ë‹¹ê¸°ìˆœì´ìµ(ë¹„ì§€ë°°)', 'ìì‚°ì´ê³„', 'ë¶€ì±„ì´ê³„',
           'ìë³¸ì´ê³„', 'ìë³¸ì´ê³„(ì§€ë°°)', 'ìë³¸ì´ê³„(ë¹„ì§€ë°°)', 'ìë³¸ê¸ˆ', 'ì˜ì—…í™œë™í˜„ê¸ˆíë¦„',
           'íˆ¬ìí™œë™í˜„ê¸ˆíë¦„', 'ì¬ë¬´í™œë™í˜„ê¸ˆíë¦„', 'ì˜ì—…ì´ìµë¥ ', 'ìˆœì´ìµë¥ ', 'ROE(%)',
           'ROA(%)', 'ë¶€ì±„ë¹„ìœ¨', 'ìë³¸ìœ ë³´ìœ¨', 'EPS(ì›)', 'PER(ë°°)']

```

```python
# header ë„£ì–´ì¤Œ
import pandas as pd

df = pd.read_csv(url,header= None) # ì›ë˜ ìˆë˜ ë°ì´í„°ê°€ headerê°€ ë˜ì–´ ì‚¬ë¼ì§€ì§€ ì•Šê²Œ header=None ì¶”ê°€

df.columns = headers # ìœ„ì— ì§€ì •í•œ headerë¥¼ column(header)ë¡œ 
df
```

```python
# ë¶„ê¸°ë³„ ì˜ì—…ì´ìµë¥  _ ì˜ì—…ì´ìµë¥ 2

# ì˜ì—…ì´ìµë¥ : ì˜ì—…ì´ìµ/ë§¤ì¶œì•¡

def toInt(string):
  return int(string.replace(',',''))


df['ì˜ì—…ì´ìµ'] = df['ì˜ì—…ì´ìµ'].apply(toInt)
df['ë§¤ì¶œì•¡'] = df['ë§¤ì¶œì•¡'].apply(toInt)

# ì˜ì—…ì´ìµë¥ 2

df['ì˜ì—…ì´ìµë¥ 2'] = round((df['ì˜ì—…ì´ìµ']/df['ë§¤ì¶œì•¡'])*100,2) #ì†Œìˆ˜ì  2ìë¦¬ ì´í•˜ ë°˜ì˜¬ë¦¼

df
df.dtypes
```

```python
# 18/19 ë…„ë„ PDR ê³„ì‚°

# PDR : ì‹œê°€ì´ì•¡ / (ì „ì²´ ì‹œì¥ ê·œëª¨(TAM) * (ì‹œì¥ì ìœ ìœ¨/100)) * 100
df2 = pd.DataFrame({'êµ¬ë¶„':['18ë…„','19ë…„'], 
                    'ì¢…ê°€':[101500,93800], 
                    'ë°œí–‰ì£¼ì‹ìˆ˜':[137292497,137292497],
                    'ì‹œê°€ì´ì•¡':[13935188445500,12878036218600],
                    'ì‹œì¥ì ìœ ìœ¨':[62.0,63.5]})                    # DataFrameìœ¼ë¡œ ë§Œë“¤ì–´ì£¼ê¸°
df2['PDR'] = df2['ì‹œê°€ì´ì•¡']/(15000000000000*(df2['ì‹œì¥ì ìœ ìœ¨']/100))*100 # 'PDR' ì¶”ê°€
df2
```

### Feature Engineering

Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`
- 5% ì´ìƒ -> `A`
- -5 ~ 5% -> `B`
- -5%ì´í•˜ -> `C`
- 10%ì´í•˜ -> `D`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

20ë…„ë„ 2ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ê°’ì€ `A`ê°€ ë‚˜ì™€ì•¼í•©ë‹ˆë‹¤.

ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ í•˜ì„¸ìš”.

```python
# Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°

# 4ë¶„ê¸°ì˜ í‰ê·  ê³„ì‚°

mean2 = df.iloc[1:, 1].sum() / 4
print('mean2:',mean2)

# ì ìˆ˜ ê³„ì‚° : í‰ê· ë§¤ì¶œì•¡ * 110%, 105%, 105-95%, 90%
for idx, i in enumerate(df['ë§¤ì¶œì•¡']):
  print(i,idx)
  if i >= (mean2*1.1):
    df['Relative Performance'][idx] = 'S'
  elif i>= (mean2*1.05):
    df['Relative Performance'][idx] = 'A'
  elif i>= (mean2*0.95):
    df['Relative Performance'][idx] = 'B'
  elif i>= (mean2*0.9):
    df['Relative Performance'][idx] = 'C'
  else:
    df['Relative Performance'][idx] = 'D'

# ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ í•„ìš”í•œ ë§¤ì¶œì•¡

print('S :', round(mean2*1.1,2))
print('A :', round(mean2*1.05,2))
print('B :', round(mean2*0.95,2))
print('C :', round(mean2*0.9,2))
print('D :', round(mean2*0.9,2))

df
```
